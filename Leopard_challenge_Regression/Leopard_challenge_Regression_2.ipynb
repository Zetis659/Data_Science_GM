**Функция гистограммы и описания для количественных признаков**

```python
def hist_box_describe_num(ds, x):
    print(ds[x].describe())
    plt.figure()
    sns.histplot(data=ds, x=ds[x], kde=True);
    plt.figure()
    sns.boxplot(data=ds, x=ds[x]);
```

```python
def hist_describe_num(ds, x):
    plt.figure()
    sns.histplot(data=ds, x=ds[x], kde=True);
    print(ds[x].value_counts())
```

### Заполняю пропуски train и test

**Создам переменные трейн и тест, которые будут иметь изначальный вид, для дальнейшего использования и сравненния результатов:**

#### Bedroom2 - median

```python
train.groupby('Bedroom2').Price.median()
```

```python
train.Bedroom2.median(), test.Bedroom2.median()
```

```python
train['Bedroom2'] = train.Bedroom2.fillna(train.Bedroom2.median())
test['Bedroom2'] = test.Bedroom2.fillna(test.Bedroom2.median())
```

#### Bathroom - median

```python
train.groupby('Bathroom').Price.median()
```

```python
train.Bathroom.median(), test.Bathroom.median()
```

```python
train['Bathroom'] = train.Bathroom.fillna(train.Bathroom.median())
test['Bathroom'] = test.Bathroom.fillna(test.Bathroom.median())
```

#### Car

```python
train.Car.median(), test.Car.median()
```
